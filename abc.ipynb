```python
from rdkit import Chem


# 带有原子编号的SMILES字符串
annotated_smiles = "CC(C)(C)OC(=O)O[C:6]([O:5][C:2]([CH3:1])([CH3:3])[CH3:4])=[O:7].[CH3:8][C:9](=[O:10])[c:11]1[cH:12][cH:13][c:14]2[nH:15][cH:16][cH:17][c:18]2[cH:19]1"

# 解析带有原子编号的SMILES字符串
mol = Chem.MolFromSmiles(annotated_smiles, sanitize=False)

# 删除原子编号
for atom in mol.GetAtoms():
    atom.ClearProp('molAtomMapNumber')

# 标准化SMILES字符串
standard_smiles = Chem.MolToSmiles(mol)

print("Standard SMILES:", standard_smiles)
```

```python
from rdkit import Chem

# 定义两个SMILES字符串
smiles1 = 'CC(=O)c1ccc2[nH]ccc2c1.CC(C)(C)OC(=O)OC(=O)OC(C)(C)C'
smiles2 = 'CC(C)(C)OC(=O)OC(=O)OC(C)(C)C.CC(=O)c1ccc2[nH]ccc2c1'

# 使用RDKit的Chem.MolFromSmiles函数将SMILES字符串转换为分子对象
mol1 = Chem.MolFromSmiles(smiles1)
mol2 = Chem.MolFromSmiles(smiles2)

# 使用Chem.MolToSmiles函数将分子对象转换为标准化的SMILES字符串
standard_smiles1 = Chem.MolToSmiles(mol1)
standard_smiles2 = Chem.MolToSmiles(mol2)

print("Standard SMILES 1:", standard_smiles1)
print("Standard SMILES 2:", standard_smiles2)
print("Are they equal?", standard_smiles1 == standard_smiles2)
```
